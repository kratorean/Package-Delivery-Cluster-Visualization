```python
#Load my geocoded data
import pandas as pd
import folium
from folium.plugins import MarkerCluster

df = pd.read_csv('C:\\Users\\Co\\Documents\\abj_ops_analysis\\geocoded_data.csv')

df_sample = df.sample(5000) # Reduce to 5000 points for better performance


def map_color(Stop_Type):
    if pd.isna(Stop_Type):
        return "gray"
    Stop_type = Stop_Type.strip().upper() #to normalize my text
    if Stop_Type == "PICKUP":
        return "blue"
    elif Stop_Type == "DELIVERY":
        return "red"
    else:
        return "gray"  #would be default for unknown values
    

m = folium.Map(location=[df_sample['Latitude'].mean(), df_sample['Longitude'].mean()], zoom_start=6)
marker_cluster = MarkerCluster().add_to(m)

# Add markers for each location
for _, row in df_sample.iterrows():
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        
        color = map_color(row['Stop_Type_Desc'])

        popup_info = f"City: {row['City_Name']}<br>Driver: {row['Orgnl Pkg Driver Name']}<br>Type: {row['Sp_Type']}"

        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=6,
            popup=popup_info,
            color=color,
            fill=True,
            fill_color=color
        ).add_to(marker_cluster)

m.save("map.html")
print("Map saved successfully!")
```
